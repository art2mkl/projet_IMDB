```python
#Remplacement des NAN par les mean() selon l'anne en cours

for i in range(1920,2021):
    #si la date en i ne contient pas que des NAN
    if df1['gross(M$)'][df1['released_year'] == i].notnull().sum() != 0:
        df1['gross(M$)'][df1['released_year'] == i] = df1['gross(M$)'][df1['released_year'] == i].fillna(df1['gross(M$)'][df1['released_year'] == i].mean())
    
    #si la date en i-1 ne contient pas que des NAN
    elif df1['gross(M$)'][df1['released_year'] == i-1].notnull().sum() != 0:
        df1['gross(M$)'][df1['released_year'] == i] = df1['gross(M$)'][df1['released_year'] == i].fillna(df1['gross(M$)'][df1['released_year'] == i-1].mean())
    
    #sinon utilise le mean() de i-2
    else:
        df1['gross(M$)'][df1['released_year'] == i] = df1['gross(M$)'][df1['released_year'] == i].fillna(df1['gross(M$)'][df1['released_year'] == i-2].mean())
    

        
        
        
```
